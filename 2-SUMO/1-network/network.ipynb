1. Extract OSM data with Osmium (bounding box)

```python
import shutil
import subprocess
from pathlib import Path

# ============================================================
# Step 1 — Extract OSM data with Osmium (bounding box)
# ============================================================

# Study area bounding box (lon_min, lat_min, lon_max, lat_max)
BBOX = "-1.571,45.800,-0.533,46.353"

INPUT_FILE = Path("input/poitou-charentes-251203.osm.pbf")
OUTPUT_FILE = Path("cda_la_rochelle.osm")

# Ensure Osmium is installed and available in PATH
if shutil.which("osmium") is None:
    raise RuntimeError(
        "Osmium is not available in your PATH. Install it first (e.g., 'osmium-tool') "
        "and make sure the 'osmium' command works in your terminal."
    )

if not INPUT_FILE.exists():
    raise FileNotFoundError(f"Input file not found: {INPUT_FILE.resolve()}")

command = [
    "osmium", "extract",
    "-b", BBOX,
    str(INPUT_FILE),
    "-o", str(OUTPUT_FILE),
]

print("[INFO] Running command:", " ".join(command))
subprocess.run(command, check=True)

print(f"[OK] Extraction completed: {OUTPUT_FILE}")
```

2. Convert OSM to SUMO network (netconvert)

```python
import shutil
import subprocess
from pathlib import Path

# ============================================================
# Step 2 — Convert OSM to SUMO network (netconvert)
# ============================================================

OSM_INPUT_FILE = Path("cda_la_rochelle.osm")
NET_OUTPUT_FILE = Path("cda_la_rochelle.net.xml")

# Ensure netconvert (SUMO) is available in PATH
if shutil.which("netconvert") is None:
    raise RuntimeError(
        "netconvert is not available in your PATH. Install SUMO and ensure 'netconvert' "
        "works in your terminal (or add SUMO/bin to PATH)."
    )

if not OSM_INPUT_FILE.exists():
    raise FileNotFoundError(f"Input OSM file not found: {OSM_INPUT_FILE.resolve()}")

command = [
    "netconvert",
    "--osm-files", str(OSM_INPUT_FILE),
    "-o", str(NET_OUTPUT_FILE),
]

print("[INFO] Running command:", " ".join(command))

try:
    subprocess.run(command, check=True)
    print(f"[OK] Conversion successful: {NET_OUTPUT_FILE}")
except subprocess.CalledProcessError as e:
    print("[ERROR] netconvert failed:", e)
    raise
```

3. Generate polygon file from OSM (polyconvert)

```python
import shutil
import subprocess
from pathlib import Path

# ============================================================
# Step 3 — Generate polygon file from OSM (polyconvert)
# ============================================================

OSM_INPUT_FILE = Path("cda_la_rochelle.osm")
NET_INPUT_FILE = Path("cda_la_rochelle.net.xml")
POLY_OUTPUT_FILE = Path("cda_la_rochelle.poly.xml")

# Ensure polyconvert (SUMO) is available in PATH
if shutil.which("polyconvert") is None:
    raise RuntimeError(
        "polyconvert is not available in your PATH. Install SUMO and ensure 'polyconvert' "
        "works in your terminal (or add SUMO/bin to PATH)."
    )

if not OSM_INPUT_FILE.exists():
    raise FileNotFoundError(f"Input OSM file not found: {OSM_INPUT_FILE.resolve()}")

if not NET_INPUT_FILE.exists():
    raise FileNotFoundError(f"Input NET file not found: {NET_INPUT_FILE.resolve()}")

command = [
    "polyconvert",
    "--osm-files", str(OSM_INPUT_FILE),
    "--net-file", str(NET_INPUT_FILE),
    "-o", str(POLY_OUTPUT_FILE),
]

print("[INFO] Running command:", " ".join(command))

try:
    subprocess.run(command, check=True)
    print(f"[OK] Polygons generated: {POLY_OUTPUT_FILE}")
except subprocess.CalledProcessError as e:
    print("[ERROR] polyconvert failed:", e)
    raise
```
